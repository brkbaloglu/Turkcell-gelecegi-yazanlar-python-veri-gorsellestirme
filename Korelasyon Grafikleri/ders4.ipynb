## Scatterplot Matrisi

```python
import seaborn as sns
```

```python
iris = sns.load_dataset("iris")
```

```python
df = iris.copy()
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
sns.pairplot(df, hue = "species")
```

```python
sns.pairplot(df, hue = "species", markers = ["o", "s", "d"])
```

```python
sns.pairplot(df, kind = "reg", hue = "species")
```
